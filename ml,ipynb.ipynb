```python
data={
    "name":"nikita",
    "age" :20,
    "city":"shimla"
}
print(data)
print(type(data))
```

```python
#json
import json
json_data=json.dumps(data)
print(json_data)
print(type(json_data))
```

```python
import sys
size_json=sys.getsizeof(json_data) #in bytes
print(size_json)
```

```python
#binary
binary_data = bytes(json_data,"utf8")
print(binary_data)
print(type(binary_data))
```

```python
size_binary=sys.getsizeof(binary_data)
print(size_binary)
```
